# How feature becomes seperable in higher dimentions

```python
import numpy as np
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

X, y = make_circles(n_samples=1000, noise=0.05, factor=0.5)
print(X.shape)
print(y.shape)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

Not Linearly Seperatable

```python
col = np.sum(np.square(X), axis=1)
Z = np.hstack((X, col.reshape(-1, 1)))
print(np.shape(Z))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(xs = Z[:, 0], ys=Z[:, 1], zs = Z[:, 2], c=y)
plt.show()
```

Now In higher dimentions data is seperable. SVMs do the same things. This is done by kernals.

### Linear Kernal

```python
from sklearn.datasets import load_iris
from sklearn.svm import SVC

iris_data = load_iris()
X = iris_data.data
y = iris_data.target
print(np.shape(X))
print(np.shape(y))
```

```python
from sklearn.utils import shuffle
X, y = shuffle(X, y, random_state=42)
```

```python
plt.scatter(X[:, 2], X[:, 3], c=y)
plt.xlabel('Petal-lenght')
plt.ylabel('Petal-width')
plt.show()
```

```python
svm_clf = SVC(kernel='linear', C=1)
svm_clf.fit(X[:, [2, 3]], y)
```

```python
x_min, x_max = X[:, 2].min() - 0.5, X[:, 2].max() + 0.5
y_min, y_max = X[:, 3].min() - 0.5, X[:, 3].max() + 0.5

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
```

```python
plt.scatter(xx.ravel(), yy.ravel())
plt.show()
```

```python
Z = svm_clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(6, 4))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)
plt.scatter(X[:, 2], X[:, 3], c=y, s=10)

support_vect = svm_clf.support_vectors_
plt.scatter(support_vect[:, 0], support_vect[:, 1], c='r', s=20, label="Support Vector")
plt.xlabel('Petal-lenght')
plt.ylabel('Petal-width')
plt.title("Linear Kernal")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

def predict(svm_clf, X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)

    svm_clf.fit(X_train, y_train)

    pred_train = svm_clf.predict(X_train)
    pred_test = svm_clf.predict(X_test)
    
    print("Accuracy")
    print("Training Set: ", accuracy_score(y_true=y_train, y_pred=pred_train))
    print("Test Set: ", accuracy_score(y_true=y_test, y_pred=pred_test))
    print("\n")
    
    print("F1-Score")
    print("Training Set: ", f1_score(y_true=y_train, y_pred=pred_train, average='macro'))
    print("Test Set: ", f1_score(y_true=y_test, y_pred=pred_test, average='macro'))
```

```python
svm_clf = SVC(kernel='linear', C=0.1)
predict(svm_clf, X, y)
```

### Polynomial Kernal

```python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000, noise=0.1)
print(X.shape)
print(y.shape)
plt.scatter(X[:, 0], X[:, 1], c=y, s=10)
plt.show()
```

```python
def plot(X, y, kernel):
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

    configurations = [
        {'C': 1, 'gamma': 1},
        {'C': 100, 'gamma': 1},
        {'C': 1, 'gamma': 10},
        {'C': 100, 'gamma': 10}
    ]
    
    fig, axes = plt.subplots(2, 2, figsize=(10, 6))
    for ax, config in zip(axes.ravel(), configurations):
        svm_clf = SVC(kernel=kernel, gamma=config['gamma'], C=config['C'], random_state=42)
        svm_clf.fit(X, y)
        Z = svm_clf.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        
        ax.contourf(xx, yy, Z, cmap=plt.cm.Paired)
        ax.scatter(X[:, 0], X[:, 1], c=y, s=5)
        support_vect = svm_clf.support_vectors_
        sv = len(support_vect)
        ax.scatter(support_vect[:, 0], support_vect[:, 1], c='r', s=10, label="Support Vector")
        ax.set_title(f"C: {config['C']}, Gamma: {config['gamma']}, No of SV: {sv}")
        ax.legend()
        
    plt.tight_layout()
    plt.suptitle(kernel)
    plt.show()
```

`C` balances the margin size and misclassification rate:  
Small C: Large margin, more misclassification allowed.  
Large C: Small margin, less misclassification allowed.  
`gamma` controls the influence range of data points:  
Small gamma: Broad influence, smooth boundaries.  
Large `gamma Results in a more complex, tightly fitting decision boundary. 

```python
plot(X, y, kernel='poly')
```

### RBF kernal

```python
plot(X, y, kernel='rbf')
```

# Complete Model Training and Hyper-Parameter Tuning
## Iris Dataset

```python
from sklearn.pipeline import Pipeline
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.utils import shuffle


iris = load_iris()
X, y = shuffle(iris.data, iris.target, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)

svm_clf = Pipeline(steps=[
    ('scaler', StandardScaler()),
    ('svc', SVC())
])

# Hyper-Parameters
param_dict = {"svc__kernel": ['linear', 'poly', 'rbf', 'sigmoid'],
             "svc__C": [0.01, 0.1, 1, 10, 100],
             "svc__gamma": [0.0001, 0.001, 0.1, 1]}

grid_search = GridSearchCV(
    estimator=svm_clf,
    param_grid=param_dict,
    cv=5, 
    verbose=3,
    n_jobs=-1,
    refit=True
)

grid_search.fit(X_train, y_train)
```

```python
print(grid_search.best_params_)
print(grid_search.best_estimator_)
print(grid_search.best_score_)
```

```python
from sklearn.metrics import classification_report

train_pred = grid_search.predict(X_train)
test_pred = grid_search.predict(X_test)

print("Classification Report: Training Set: ")
print(classification_report(y_true=y_train, y_pred=train_pred))
print("Classification Report: Test Set: ")
print(classification_report(y_true=y_test, y_pred=test_pred))
```

# LinearSVM
* LinearSVM is faster than SVMs. It requires no kernel trick

```python
from sklearn.svm import LinearSVC

model = LinearSVC(C=10, loss='hinge', max_iter=1000)
model.fit(X_train, y_train)

train_pred = model.predict(X_train)
test_pred = model.predict(X_test)

print("Classification Report: Training Set: ")
print(classification_report(y_true=y_train, y_pred=train_pred))
print("Classification Report: Test Set: ")
print(classification_report(y_true=y_test, y_pred=test_pred))
```

# Regression
* `SVR`

```python
X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel()
plt.scatter(X, y)
plt.show()
```

```python
from sklearn.svm import SVR
svrs = [SVR(kernel='linear'), SVR(kernel='poly'), SVR(kernel='rbf', epsilon=0.2)]

for svr in svrs:
    svr.fit(X, y)
```

```python
plt.scatter(X, y, s=10)
kernels = ['linear', 'poly', 'rbf']
c = ['g', 'r', 'b']
for i in range(len(svrs)):
    plt.plot(X, svrs[i].predict(X), c=c[i], label=kernels[i] + " kernel")
plt.legend()
plt.show()
```

`LinearSVR`

```python
X = np.linspace(-20, 20, 200).reshape(-1, 1)
y = np.square(X) + np.random.normal(loc=0, scale=50, size=X.shape)

plt.scatter(X, y)
plt.show()
```

```python
from sklearn.svm import LinearSVR
svr = LinearSVR(C=100, epsilon=0.5)   # Not suitable for non linear data
svr.fit(X, y)
plt.scatter(X, y)
plt.plot(X, svr.predict(X))
plt.show()
```

```python
from sklearn.svm import SVR
svr = SVR(kernel='poly', C=100, degree=2, epsilon=1)
svr.fit(X, y)
plt.scatter(X, y)
plt.plot(X, svr.predict(X))
plt.title("Poly")
plt.show()
```

```python
svr = SVR(kernel='rbf', C=100, epsilon=0.1)
svr.fit(X, y)
plt.scatter(X, y)
plt.plot(X, svr.predict(X))
plt.title("rbf")
plt.show()
```

# California Housing Data

```python
import pandas as pd
housing = pd.read_csv('data/housing.csv')
housing.head()
```

```python
housing.isnull().sum()
```

```python
housing.info()
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)
print(train_set.shape)
print(test_set.shape)
```

```python
X_train = train_set.drop('median_house_value', axis=1)
y_train = train_set['median_house_value']
X_test = test_set.drop('median_house_value', axis=1)
y_test = test_set['median_house_value']

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

num_attribs = ['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income']
cat_attribs = ['ocean_proximity']

num_pipeline = make_pipeline(
    SimpleImputer(strategy='mean'),
    StandardScaler()
)

cat_pipeline = make_pipeline(
    SimpleImputer(strategy='most_frequent'),
    OneHotEncoder(handle_unknown='ignore')
)

preprocessing = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', cat_pipeline, cat_attribs)
])

prepared_data = preprocessing.fit_transform(X_train)
print(preprocessing.get_feature_names_out())
```

```python
df = pd.DataFrame(prepared_data, columns=preprocessing.get_feature_names_out(), index=X_train.index)
print(df.shape)
df.head()
```

```python
from sklearn.svm import SVR
svm_reg = Pipeline(steps=[
    ('preprocessing', preprocessing),
    ('reg', SVR(kernel='linear', gamma=0.3, C=10))])

svm_reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

train_pred = svm_reg.predict(X_train)
test_pred = svm_reg.predict(X_test)

print("Training RMSE: ", np.sqrt(mean_squared_error(y_pred=train_pred, y_true=y_train)))
print("Test RMSE: ", np.sqrt(mean_squared_error(y_pred=test_pred, y_true=y_test)))
```
